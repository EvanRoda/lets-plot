# Superscript in Lets-Plot

```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

#### 1. Default Behaviour

```python
# data source: https://physics.nist.gov/cuu/Constants/Table/allascii.txt
charged_leptons_data = {
    "name": ["electron", "muon", "tau"],
    "mass_kg": [9.1093837015e-31, 1.883531627e-28, 3.16754e-27],
}
```

```python
ggplot(charged_leptons_data) + \
    geom_lollipop(aes("name", "mass_kg"))
```

#### 2. Formatting

```python
def get_plot(tooltips=None):
    x = list(range(1, 11))
    y = [i * 10**(-5) for i in x]
    return ggplot() + \
        geom_bar(aes(x=x, y=y), stat='identity', tooltips=tooltips) + \
        ggtitle("Default" if tooltips is None else "Formatted tooltips")

gggrid([
    get_plot(),
    get_plot(layer_tooltips().format("@y", ".3~e").line("@|@y"))
], ncol=2)
```
