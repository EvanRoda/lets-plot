```python
from lets_plot import *
LetsPlot.setup_html()

from lets_plot.plot.coord import coord_polar
```

```python
hours = list(map(lambda v: '{:02d}:00'.format(v), range(12)))
p = ggplot() \
    + geom_point(aes(x=hours)) \
    + theme_grey()
```

# Default
Default padding may cause tick labels overlapping in polar coordinate system:

```python
gggrid([
    p,
    p + coord_polar(),
])
```

# `theme(panel_padding=[0, 35, 0, 35])`
The `panel_padding` parameter can be used to create space between the axis and the panel content:

```python
padding = [10, 35, 10, 35]
gggrid([
    p + theme(panel_padding=padding) + ggtitle('padding=' + str(padding)),
    p + theme(panel_padding=padding) + coord_polar() + ggtitle('padding=' + str(padding)),
], ncol=2)
```

# `panel_padding` and `coord_fixed()`
`panel_padding` also works with `coord_fixed()` and all axis.

```python
padding = 40
p \
    + scale_x_continuous(position='both') \
    + scale_y_continuous(position='both') \
    + coord_fixed() \
    + ggtitle('coord_fixed() + padding=' + str(padding)) \
    + theme(panel_padding=padding) 
```
